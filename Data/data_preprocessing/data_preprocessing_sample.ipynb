```python
import pandas as pd
```

```python
# 브이월드 '토지소유정보(충청남도)' 데이터를 'land_sample.csv'에 미리 저장한 상태
# '토지소유정보' 데이터를 df 변수에 불러오기

df = pd.read_csv('sample_data/land_sample.csv')
```

```python
# df를 '개인소유 농지'로 필터링하여 df_land에 저장

df_land = df.loc[(df['A4'] == '토지대장') & (df['A8'] == '개인') & (df['A20'].isin(['전','답','과수원']))]
```

```python
# df_land에 '주소' 컬럼 추가
# 'A2': 법정동명
# 'A5': 지번

df_land['주소'] = df_land['A2'] + ' ' + df_land['A5']
```

```python
# 필요없는 컬럼 제거

drop_col = ['A2','A3','A4','A5','A6','A7','A8','A9','A10','A11','A12','A13','A14','A15','A16','A17','A18','A19','A21','A23','A24']
df_land = df_land.drop(columns=drop_col)
```

```python
# '토지이용정보' 컬럼을 통해 실제로 지목이 어떻게 사용되고 있는지 확인 가능
# '토지이용정보' 컬럼을 사전에 농지 지목 (전, 답, 과수원)으로 전처리하여 'sample.csv'로 저장하였음
# df_info 변수에 '충남_토지특성정보' 불러오기

df_info = pd.read_csv('sample_data/info_sample.csv')
```

```python
# 전체 토지 중 실제로 농지로 사용되고 있는 주소로 필터링

df_merged = pd.merge(df_land, df_info,on='주소',how='inner')
```

```python
# 건축물 대장 데이터를 사전에 'building_sample.csv'로 저장한 상태
# '건축물 대장 데이터'를 df_building 변수에 불러오기

df_building = pd.read_csv('sample_data/building_sample.csv')
```

```python
# 허가받은 건축물이 있는 주소를 리스트로 만들어, 허가 받은 건축물이 있는 농지는 제외

building_list = df_building['주소'].tolist()
pre_data = df_merged[~df_merged['주소'].isin(building_list)]
```

```python
# 데이터 인덱스 초기화 및 컬럼 삭제, 이름 설정

pre_data.reset_index(inplace=True)
pre_data.drop(columns=['index','Unnamed: 0.1','Unnamed: 0_x','Unnamed: 0_y','토지이용상황'], inplace=True)
pre_data.rename(columns={'A0':'고유번호','A1':'법정동코드','A20':'지목','A22':'토지면적'},inplace=True)
pre_data
```

```python
pre_data.to_csv('preprocessed_data/pre_data.csv')
```
