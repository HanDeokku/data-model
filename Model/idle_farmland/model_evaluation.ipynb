```python
import pandas as pd
df = pd.read_csv('data_100.csv')
```

```python
from shapely.geometry import Polygon

def string_to_polygon(polygon_str):
    coordinates = polygon_str.replace('POLYGON ((', '').replace('))', '').replace(')', '').replace('(', '').split(', ')
    coordinates = [tuple(map(float, coord.split())) for coord in coordinates]
    return Polygon(coordinates)
```

```python
import cv2
import numpy as np

def check_change(path, ply_list):
    change_map = cv2.imread(path, cv2.IMREAD_GRAYSCALE)

    # 폴리곤 좌표를 지정합니다 (예시로 임의의 폴리곤 좌표를 사용합니다)
    polygon_coords = np.array(ply_list, dtype=np.float32)
    polygon_coords = polygon_coords.astype(np.int32)

    polygon_mask = np.zeros_like(change_map)

    # 폴리곤 내부를 흰색(255)으로 채웁니다
    cv2.fillPoly(polygon_mask, [polygon_coords], 255)

    # 변화 맵에 폴리곤 마스크를 적용하여 해당 영역의 변화를 추출합니다
    polygon_change = cv2.bitwise_and(change_map, polygon_mask)

    # 폴리곤 내부의 흰색 픽셀 수 계산
    num_white_pixels_polygon = np.count_nonzero(polygon_change == 255)

    # 폴리곤 영역 전체 픽셀 수 계산
    total_pixels_polygon = np.count_nonzero(polygon_mask == 255)

    # 변화 비율 계산
    if total_pixels_polygon > 0:
        change_ratio_polygon = num_white_pixels_polygon / total_pixels_polygon
    else:
        change_ratio_polygon = 0.0

    return change_ratio_polygon
```

```python
빈땅예측 = []
유휴농지예측 = []
```

```python
def get_filenames_without_extension(folder_path):
    filenames = []
    for filename in os.listdir(folder_path):
        if os.path.isfile(os.path.join(folder_path, filename)):
            name, _ = os.path.splitext(filename)
            filenames.append(name)
    return filenames

# 예시 사용
folder_path = 'output1'  # 폴더 경로를 여기에 입력하세요
file_list = get_filenames_without_extension(folder_path)
print(file_list)
```

```python
len(file_list)
```

```python
from ctypes import addressof
path1 = 'output1'
path2 = 'output2'

for address in file_list:
    address = unicodedata.normalize('NFKC', address)
    file_path1 = os.path.join(path1, address+'.jpg')
    file_path2 = os.path.join(path2, address+'.jpg')
    polygon = df.loc[df['주소'] == address, 'pixel_polygon'].values[0]
    ply = string_to_polygon(polygon)
    ply_list = list(ply.exterior.coords)
    score1 = check_change(file_path1, ply_list)
    score2 = check_change(file_path2, ply_list)
    score = (score1 + score2) / 2
    print(f'주소: {address} / score1 : {score1} / score2 : {score1} / score : {score}')
    if score >= 0.06:
        빈땅예측.append(address)
    else:
        유휴농지예측.append(address)
```

```python
folder_path = 'image_data/cultivate'  # 여기에 대상 폴더 경로를 넣으세요.
빈땅실제 = []

for foldername, subfolders, filenames in os.walk(folder_path):
    for subfolder in subfolders:
        # 하위 폴더의 전체 경로를 얻고 싶으면 다음 줄을 사용하세요.
        # subfolder_path = os.path.join(foldername, subfolder)
        name = unicodedata.normalize('NFKC', subfolder)
        # 하위 폴더 이름을 리스트에 추가
        빈땅실제.append(name)

# 결과 출력
print(빈땅실제)
```

```python
folder_path = 'image_data/idle'  # 여기에 대상 폴더 경로를 넣으세요.
유휴농지실제 = []

for foldername, subfolders, filenames in os.walk(folder_path):
    for subfolder in subfolders:
        # 하위 폴더의 전체 경로를 얻고 싶으면 다음 줄을 사용하세요.
        # subfolder_path = os.path.join(foldername, subfolder)
        name = unicodedata.normalize('NFKC', subfolder)
        # 하위 폴더 이름을 리스트에 추가
        유휴농지실제.append(name)

# 결과 출력
print(유휴농지실제)
```

```python
len(유휴농지실제), len(빈땅실제)
```

```python
# True Positive (TP): 실제 유휴 농지 중 예측도 유휴 농지인 경우의 수
TP = len([item for item in 빈땅실제 if item in 빈땅예측])

# True Negative (TN): 실제 빈땅 중 예측도 빈땅인 경우의 수
TN = len([item for item in 유휴농지실제 if item in 유휴농지예측])

# False Positive (FP): 실제 빈땅 중 예측이 유휴 농지인 경우의 수
FP = len([item for item in 유휴농지실제 if item in 빈땅예측])

# False Negative (FN): 실제 유휴 농지 중 예측이 빈땅인 경우의 수
FN = len([item for item in 빈땅실제 if item in 유휴농지예측])

# Confusion Matrix 출력
confusion_matrix = {
    'True Positive (TP)': TP,
    'True Negative (TN)': TN,
    'False Positive (FP)': FP,
    'False Negative (FN)': FN
}

print(confusion_matrix)
```

```python
# 정확도 (Accuracy) 계산
accuracy = (TP + TN) / (TP + TN + FP + FN)

# 정밀도 (Precision) 계산
precision = TP / (TP + FP) if (TP + FP) != 0 else 0

# 재현율 (Recall) 계산
recall = TP / (TP + FN) if (TP + FN) != 0 else 0

# F1-score 계산
f1_score = 2 * (precision * recall) / (precision + recall) if (precision + recall) != 0 else 0

# 결과 출력
print(f"정확도 (Accuracy): {accuracy:.2f}")
print(f"정밀도 (Precision): {precision:.2f}")
print(f"재현율 (Recall): {recall:.2f}")
print(f"F1-score: {f1_score:.2f}")
```
